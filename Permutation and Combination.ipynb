# Example: Permutation:
Permutation refers to the arrangement of objects in a specific order
   Q1, suppose you want to form a delegation with a manager, empolye, and
             secretary from the same pool of 8 candidates

P (n, r) =n! / (n−r)!
P (8,3) =8! / (8−3)! =8! /5!
= 8×7×6×5×4×3×2×1 /5×4×3×2×1 =336

# Example: Combination
Q2 Suppose there are 8 candidates, and you need to select a committee of 3 people from this pool. However, the order in which you select them doesn't matter.
combination formula C (n, r) =n! /r! *(n−r)!  
C (8,3) =8! / 3! (8−3)! = 8! / 3! *5!
=8×7×6 / 3×2×1= 56


```python
import itertools

# List of candidate names
candidates = [
    "Alice", "Bob", "Charlie", "David", "Eve",
    "Frank", "Grace", "Hannah"
]

# Generate all combinations of 3 candidates
committee_combinations = itertools.combinations(candidates, 3)

# Generate all permutations of 3 candidates for manager, empolye, and secretary
delegation_permutations = itertools.permutations(candidates, 3)

# Print the permutations
print("Possible delegations with manager, empolye, and secretary:")
for idx, delegation in enumerate(delegation_permutations, 1):
    print(f"{idx}: manager: {delegation[0]}, empolye: {delegation[1]}, Secretary: {delegation[2]}")

# Print the combinations
print("Possible committees of 3 people:")
for idx, committee in enumerate(committee_combinations, 1):
    print(f"{idx}: {', '.join(committee)}")

# Count the total number of combinations
total_permutations = sum(1 for _ in itertools.permutations(candidates, 3))
print("\nThe total number of possible delegation_permutations:", total_permutations)

# Count the total number of combinations
total_combinations = sum(1 for _ in itertools.combinations(candidates, 3))
print("\nThe total number of possible committees:", total_combinations)
```
